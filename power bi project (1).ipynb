```python
Power BI Project: COVID-19 Dashboard
Project Overview
The COVID-19 Dashboard is a comprehensive and interactive visualization tool created using Power BI. The dashboard provides insights into the global impact of the COVID-19 pandemic through various visualizations, including trends over time, regional data, weekly moving averages, and country-specific case counts.

Key Insight:
Covid cases and deaths were increasing till 2021. From 2022 onwards,
Covid was at its peak in 2021.
From 2022 onwards intensity of covid started reducing.
USA, India and Brazil were affected by covid the most.
USA in AMRO region was affected the most by COVID.
Features
Time-Series Analysis:

Trends of daily and cumulative COVID-19 cases and deaths.
Weekly moving averages for a smoother trend representation.
Regional Data:

Breakdown of COVID-19 data by WHO regions.
Country-specific case counts within each region.
Geographical Insights:

Interactive map chart showing case distribution globally.
Country-level drill-down for detailed analysis.
Data Visualization:

Bar charts and line charts for clear representation.
Hierarchies for seamless navigation between regions and countries.
Dashboard:

A consolidated view of all key metrics and insights.
User-friendly interface with interactive filters and slicers.
Tools and Technologies
Power BI Desktop: For creating the dashboard and visualizations.
Data Sources: COVID-19 datasets (publicly available data from WHO or Johns Hopkins University).
DAX (Data Analysis Expressions): For creating calculated measures and custom metrics.
Power Query: For data cleaning and transformation.
Objectives
Provide a comprehensive analysis of the COVID-19 pandemic.
Enable users to track trends and gain regional insights.
Support data-driven decision-making through interactive visualizations.
Key Metrics
Total confirmed cases.
Total deaths.
Case fatality ratio.
Daily new cases.
Weekly moving averages.
Regional and country-specific distributions.
How to Use the Dashboard
Open the dashboard in Power BI Desktop or Power BI Service.
Use filters and slicers to select specific time periods, regions, or countries.
Hover over visualizations for detailed tooltips.
Click on map regions or charts to drill down for more granular data.
About
No description, website, or topics provided.
Resources
 Readme
 Activity
Stars
 1 star
Watchers
 1 watching
Forks
 0 forks
Report repository
Releases
No releases published
Packages
No packages published
Footer
```
